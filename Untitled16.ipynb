```python
import pandas as pd
store = pd.read_csv('/content/Stores.csv')
store
```

```python
store_area = list(store['Store_Area'])
store_ID = list(store['Store ID '])
items_available = list(store['Items_Available'])
daily_customer_count = list(store['Daily_Customer_Count'])
store_sales = list(store['Store_Sales'])
```

```python
number_of_sales = len(store_sales)
number_of_sales
```

```python
max_sales = max(store_sales)
max_sales
```

```python
min_sales = min(store_sales)
min_sales
```

```python
average_sales = sum(store_sales)/number_of_sales
print(int(average_sales))
```

```python
import numpy as np
```

```python
average = np.mean(store_sales)
average
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(store_sales, edgecolor = 'black')
```

```python
plt.boxplot(store_sales)
plt.title('Box plot showing Store Sales')
plt.ylabel('Store Sales')
```
